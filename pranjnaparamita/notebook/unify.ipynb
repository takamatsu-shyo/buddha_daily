```python
import os
```

```python
dir_list = os.listdir("./content")
```

```python
!mkdir unified
```

```python
for d in dir_list:
    output_file_path = os.path.join("./unified", d)
    print(output_file_path)
    
    content_sub_dir = os.path.join("content",d)
    
    unified_content = ""

    files = os.listdir(content_sub_dir)
    if len(files) != 1:
        #for f in files[:2]:
        for f in files:
            file_path = os.path.join(content_sub_dir, f)
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                unified_content += content + "\n"  # Adding newline for separation
            
    else:
        file_path = os.path.join(content_sub_dir, str(files)[2:-2])
        with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                unified_content += content + "\n"  # Adding newline for separation

    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write(unified_content)
            
```
