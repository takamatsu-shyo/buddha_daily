```python
import os
import re
```

```python
data_dir = "./data"
```

```python
output_dir = "./content"
```

```python
dirs = os.listdir(data_dir)
```

```python
dirs[0]
```

```python
file_path_list = []

for d in dirs:
    data_sub_dir = os.path.join(data_dir, d)
    out_sub_dir = os.path.join(output_dir, d)

    print(out_sub_dir)
    os.makedirs(out_sub_dir, exist_ok=True)
    
    files = os.listdir(data_sub_dir)
    if len(files) != 1:
        #for f in files[:2]:
        for f in files:
            file_path = os.path.join(data_sub_dir, f)
            file_path_list.append(file_path)
    else:
        file_path = os.path.join(data_sub_dir, str(files)[2:-2])
        file_path_list.append(file_path)
    
```

```python
file_path_list
```

```python
def strip(input_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        content = file.read()
    
    # Extracting the relevant section using regex
    match = re.search(r'三藏法師玄奘奉　詔譯(.*?){{footer', content, re.DOTALL)
    
    if match:
        extracted_content = match.group(1)
        
        # Remove unwanted lines
        cleaned_content = []
        for line in extracted_content.splitlines():
            if '三藏法師玄奘奉　詔譯' not in line and '{{footer' not in line:
                cleaned_content.append(line)
    
    return cleaned_content
```

```python
for file in file_path_list:
    out_file_path = file.replace("data","content")
    print(out_file_path)
    content = strip(file)

    with open(out_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write('\n'.join(content))
```
