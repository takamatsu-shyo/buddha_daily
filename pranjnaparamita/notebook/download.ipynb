```python
def gen_url(d, num, short=False):
    if short:
        url = "https://zh.wikisource.org/wiki/Special:Export/" + d
    else:
        url = "https://zh.wikisource.org/wiki/Special:Export/" + d + str(num).zfill(3)
    return url
```

```python
zipped = list(zip(dirs, ranges))
```

```python
import subprocess
import time
```

```python
for d,r in zipped:

    dl_dir = os.path.join("data", str(d))
    print(dl_dir)
    
    os.makedirs(dl_dir)
    
    #for index in range(ranges[r][0], ranges[r][0] + 1):
    for index in range(ranges[r][0], ranges[r][1] + 1):

        if d in ("8","9","10"):
            dl_url =  gen_url(dirs[d], str(index), short=True)
        else:
            dl_url = gen_url(dirs[d], str(index))

        print(dl_url)

        subprocess.run(["wget", "-P", dl_dir, dl_url])

        time.sleep(1)
```
