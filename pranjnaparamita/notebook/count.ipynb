```python
import collections
import re
import os
import matplotlib.pyplot as plt
```

```python

# Dictionary to hold characters from each file
file_characters = {}

# Read and collect Chinese characters from each file
for filename in file_list:
    if os.path.exists(filename):
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()
            chinese_characters = set(re.findall(r'[\u4e00-\u9fff]', text))  # Use set for uniqueness
            file_characters[filename] = chinese_characters
    else:
        print(f"Warning: {filename} does not exist.")

# Iterate over each file and find unique characters
unique_characters_per_file = {}

for filename, characters in file_characters.items():
    # Find characters that are not in any other file
    other_characters = set().union(*(v for k, v in file_characters.items() if k != filename))
    unique_chars = characters - other_characters
    unique_characters_per_file[filename] = unique_chars

# Print the top 10 unique characters for each file
for filename, unique_chars in unique_characters_per_file.items():
    # Count occurrences of unique characters within the file
    char_counter = collections.Counter(unique_chars)
    top_unique_chars = char_counter.most_common(100)  # Get the top 10 unique characters
    print(f"Top 100 unique characters in {filename}:")
    for char, count in top_unique_chars:
        print(f'    {char}: {count} occurrences')  # Note: count will always be 1 for unique chars
    print()  # Blank line for better separation
```

```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, leaves_list
import seaborn as sns
import matplotlib.pyplot as plt

# Load counts: CSV with index=word, columns=book1,book2,...
counts = pd.read_csv('word_counts.csv', index_col=0)

# Convert to probability distributions with add-alpha smoothing
alpha = 0.1
probs = (counts + alpha).div((counts + alpha).sum(axis=0), axis=1)

# Jensen-Shannon divergence helper
def jensen_shannon(p, q):
    m = 0.5 * (p + q)
    # use base-2 log for JSD in bits
    kl = lambda a, b: np.sum(np.where(a > 0, a * np.log2(a / b), 0.0))
    return 0.5 * (kl(p, m) + kl(q, m))

# Pairwise JSD matrix
books = probs.columns.tolist()
n = len(books)
jsd_mat = np.zeros((n, n))
for i in range(n):
    for j in range(i, n):
        d = jensen_shannon(probs.iloc[:, i].values, probs.iloc[:, j].values)
        jsd_mat[i, j] = jsd_mat[j, i] = d

jsd_df = pd.DataFrame(jsd_mat, index=books, columns=books)

# Heatmap with clustering (order by hierarchical clustering)
link = linkage(squareform(jsd_mat), method='average')
order = leaves_list(link)
ordered = jsd_df.iloc[order, :].iloc[:, order]

plt.figure(figsize=(8, 6))
sns.heatmap(ordered, cmap='magma', annot=True, fmt='.3f')
plt.title('Jensen–Shannon Divergence between books')
plt.savefig("jsd.png", dpi=300)
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.special import gammaln

counts = pd.read_csv('word_counts.csv', index_col=0)  # rows=word, cols=books
# prior: Dirichlet with alpha = 0.01 * corpus_freq (informative prior)
corpus = counts.sum(axis=1)
alpha0 = 0.01
alpha = alpha0 * (corpus / corpus.sum())

# Add prior and get posterior counts
post = counts.add(alpha, axis=0)

# For each book vs rest compute log-odds ratio with variance
books = counts.columns
results = {}
for b in books:
    a = post[b]
    b_rest = post.sum(axis=1) - a
    n_a = a.sum()
    n_b = b_rest.sum()
    # log odds
    logod = np.log(a / (n_a - a)) - np.log(b_rest / (n_b - b_rest))
    # variance (approx)
    var = 1/(a) + 1/(n_a - a) + 1/(b_rest) + 1/(n_b - b_rest)
    z = logod / np.sqrt(var)
    df = pd.DataFrame({'logod': logod, 'z': z})
    df = df.replace([np.inf, -np.inf], 0).fillna(0)
    results[b] = df.sort_values('z', ascending=False)
```
